> Standardization of Numerical Variables



```python
df[df_numerical] = df[df_numerical].apply(lambda x: (x - x.mean()) / x.std())

df.head()
```



Standardization, also known as z-score normalization, is a technique in data processing where numerical features are transformed to have a mean of 0 and a standard deviation of 1. This transformation helps to center the data and ensure it follows a standard normal distribution, making it easier to compare values on the same scale.





>Conversion of Categorical Data into Dummy Variables



```python
df_encoding=pd.get_dummies(df,columns=['SKU','BU','Brand','Model'] ,drop_first=False)

dummy_cols = [col for col in df_encoding.columns if col not in df.columns]  # Get the new columns created by get_dummies
df_encoding[dummy_cols] = df_encoding[dummy_cols].astype(int)  # Convert only these columns to int
```

```python
df_encoding.head()
```

Converting categorical data into dummy variables, often done using one-hot encoding, is essential in data preprocessing for several reasons, especially when working with machine learning algorithms.

Machine Learning Compatibility: Many machine learning models, particularly linear models, decision trees, and neural networks, expect numerical input. Categorical data, if not transformed, cant be processed directly by these models. One-hot encoding converts categories into numerical form without introducing unintended ordinal relationships.
